```python
from utils import load_flights
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
#ValueError: Mime type rendering requires nbformat>=4.2.0 but it is not installed
```

```python
# What is the percentage of delayed flights?
flights["Delayed"] = flights["DepDelay"] > 0
flights["Delayed"].mean()
```

```python
delayByCarrier = flights.groupby("UniqueCarrier")[
    "Delayed"].mean().sort_values(ascending=False) * 100

plt.figure(figsize=(10, 6))
sns.barplot(x=delayByCarrier.index, y=delayByCarrier.values)
plt.xlabel("Airline")
plt.ylabel("Percentage of Delayed Flights")
plt.title("Percentage of Delayed Flights by Airline")
for i, v in enumerate(delayByCarrier):
    plt.text(i, v + 0.5, str(round(v, 1)), ha="center", fontsize=8)
plt.savefig("../plots/delayByCarrier.png")
plt.show()
```

```python
# Interactive plot
df = delayByCarrier.reset_index()
df.columns = ["airline", "percentage"]

# plot a bar chart using plotly
fig = px.bar(df, x="airline", y="percentage",
             title="Percentage of delayed flights by airline")
fig.update_layout(xaxis_tickangle=-45)
fig.show()
```

```python
delayByMonth = flights.groupby("Month")["Delayed"].mean().sort_values() * 100
months = ["January", "February", "March", "April", "May", "June", "July", "August", "September", "October", "November", "December"]

plt.figure(figsize=(10, 6))
sns.barplot(x=delayByMonth.index, y=delayByMonth.values)
plt.xlabel("Month")
plt.ylabel("Percentage of Delayed Flights")
plt.title("Percentage of Delayed Flights by Month")
plt.xticks(range(12), months)
plt.xticks(rotation=30)
for i, v in enumerate(delayByMonth):
    plt.text(i, v + 0.5, str(round(v, 1)), ha="center", fontsize=8)
plt.savefig("../plots/delayByMonth.png")
plt.show()
```

```python
delayByDay = flights.groupby("DayOfWeek")["Delayed"].mean().sort_values() * 100
days = ["Monday", "Tuesday", "Wednesday", "Thursday", "Friday", "Saturday", "Sunday"]

plt.figure(figsize=(10,6))
sns.barplot(x=delayByDay.index, y=delayByDay.values)
plt.xlabel("Day of Week")
plt.ylabel("Percentage of Delayed Flights")
plt.title("Percentage of Delayed Flights by Day of Week")
plt.xticks(range(7), days)
plt.xticks(rotation=30)
for i, v in enumerate(delayByDay):
    plt.text(i, v + 0.5, str(round(v, 1)), ha="center", fontsize=15)
plt.savefig("../plots/delayByDay.png")
plt.show()
```

```python
# Calculate the average departure delay time for each airline and sort them in descending order
avg_delay_by_airline = flights.groupby("UniqueCarrier")[
    "DepDelay"].mean().sort_values(ascending=False)

# Plot a boxplot to show the distribution of departure delay time for each airline
plt.figure(figsize=(20, 20))
sns.boxplot(y="UniqueCarrier", x="DepDelay", data=flights)
plt.xlabel("Airline")
plt.ylabel("Departure Delay Time (minutes)")
plt.title("Departure Delay Time by Airline")
plt.savefig("../plots/delayByCarrierDistribution.png")
plt.show()
```

```python
corrMatrix = flights[flights.columns[flights.dtypes != "object"]].corr()

plt.figure(figsize=(20, 20))
sns.heatmap(corrMatrix, annot=True, cmap="coolwarm")
plt.title("Correlation Matrix")
plt.savefig("../plots/correlation_matrix.png")
plt.show()
```

```python
seasons = {1: 'Winter', 2: 'Winter', 3: 'Spring', 4: 'Spring', 5: 'Spring', 6: 'Summer', 7: 'Summer', 8: 'Summer', 9: 'Fall', 10: 'Fall', 11: 'Fall', 12: 'Winter'}

flights['Season'] = flights['Month'].map(seasons)


# Calculate the percentage of delayed flights by season
delayed = df[df['ArrDelay'] > 0] # Filter only delayed flights
delayed_by_season = delayed.groupby('Season')['FlightNum'].count() # Count the number of delayed flights by season
total_by_season = df.groupby('Season')['FlightNum'].count() # Count the total number of flights by season
percent_delayed_by_season = delayed_by_season / total_by_season * 100 # Calculate the percentage of delayed flights by season

plt.figure(figsize=(8,8))
sns.barplot(x=percent_delayed_by_season.index,y=percent_delayed_by_season.values)
plt.title('Percentage of Delayed Flights by Season')
for i, v in enumerate(percent_delayed_by_season):
    plt.text(i, v + 0.5, str(round(v, 1)), ha="center", fontsize=15)
plt.savefig("../plots/delayBySeason.png")
plt.show()
```
